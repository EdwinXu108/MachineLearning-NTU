# 相关性分析

```python
features = train_data.iloc[:, 38: -1]
display(features)
```

```python
correlations = features.corrwith(train_data.iloc[:, -1])
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=correlations.index, y=correlations.values)
plt.xticks(rotation=90)
plt.xlabel('Columns')
plt.ylabel('Correlation')
plt.title('Correlation between features and tested positive')
plt.show()
```

```python
sorted_correlations = abs(correlations).sort_values(ascending=False)
sorted_correlations
```

```python
selected_features =  sorted_correlations[sorted_correlations >= 0.8]
selected_features
```

```python
selected_column_index = []
for i in selected_features.index:
    selected_column_index.append(train_data.columns.tolist().index(i))
selected_column_index
```
